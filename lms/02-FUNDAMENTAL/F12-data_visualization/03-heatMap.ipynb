# heatMap
- Heatmap은 방대한 량의 데이터와 현상을 수치에 따른 색상으로 나타내는 것으로, 데이터 차원에 대한 제한은 없으나 모두 2차원으로 시각화하여 표현합니다.
- pivot: pivot이란 어떤 축, 점을 기준으로 바꾸다란 뜻입니다. 데이터 표를 재배치 할때도 pivot이라는 단어를 사용합니다

```python
# libary import
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd 
import seaborn as sns

import os

# font set
font_path = '/home/aiffel0042/anaconda3/envs/aiffel/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)

# data import 

csv_path = "./flights.csv"
data = pd.read_csv(csv_path)  # 판다스 csv 읽기 메서드로 읽어오기
flights = pd.DataFrame(data) # 판다스 데이터 프레임 객체로 ...
```

## pivot

```python
#pivot 
pivot = flights.pivot(index='year', columns='month', values='passengers')
pivot
```

```python
##
sns.heatmap(pivot)
```

```python
## option add
sns.heatmap(pivot, linewidths=.2, annot=True, fmt="d")
```

```python
sns.heatmap(pivot, cmap="YlGnBu")
```
