# 이미지 색, 공간 변경 및 필터 / 이미지 저장

## library import

```python

from PIL import Image
from PIL import ImageEnhance # 대비 모듈
import matplotlib.pyplot as plt
import os

```

## orginal img upload

```python
## #img_path = os.getenv('HOME')+'/aiffel/AIFFEL_LSG/lms/02-FUNDAMENTAL/F18-python_img/imgs/00-t.png'
img_path = './imgs/00-t.png'
im = Image.open(img_path)                # original 이미지
```

## 이미지 효과 _Contrast

- enh = ImageEnhance.Contrast(im) : 대비(Contrast)를 하기 위한 준비 작업으로 Contrast 클래스 객체 안에 im을 넣는 작업입니다.
- enh = enh.enhance(1.9) : 기존 대비보다 90%를 높여주었습니다. 대비를 높이면 위의 결과 이미지와 같이 색과 색의 구분이 더욱 뚜렷하게 나타납니다. 대비를 낮추려면 1 아래(예. 0.x)로 값을 인자

```python

enh = ImageEnhance.Contrast(im)    # enhanced contrast 이미지
enh = enh.enhance(1.9)

# 두 이미지를 그려 봅시다
f=plt.figure()
im1 = f.add_subplot(1,2,1)
im1.imshow(im)
im1.set_title("original")

im2 = f.add_subplot(1,2,2)
im2.imshow(enh)
im2.set_title("enhanced contrast")
```

## 이미지 필터

- 필터 종류
- BLUR : 이미지를 흐리게 만드는 필터입니다.
- EDGE_ENHANCE : 윤곽을 뚜렷하게 해주는 필터입니다.
- EMBOSS : 원본 이미지의 명암 경계에 따라 이미지의 각 픽셀을 밝은 영역 또는 어두운 영역으로 대체하는 필터입니다.
- FIND_EDGES : 윤곽만 표시해주는 필터입니다.
- SHARPEN : 경계선들을 더욱 날카롭게 해주어 선명도를 높이는 필터입니다.

```python
from PIL import ImageFilter

filtered_image = im.filter(ImageFilter.FIND_EDGES)  # BLUR filter 이미지

f=plt.figure()
im1 = f.add_subplot(1,2,1)
im1.imshow(im)
im1.set_title("original")

im2 = f.add_subplot(1,2,2)
im2.imshow(filtered_image)
im2.set_title("BLUR filter")
```
