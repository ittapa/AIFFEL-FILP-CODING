```python
def image_resize(image, height): # ì´ë¯¸ì§€ ê°ì²´ , ì„¸ë¡œ ì‚¬ì´ì¦ˆ
    
    if height == 300: # 300 ì´ë©´  800 300
        return image.resize((800, 300))
    else: # ì•„ë‹ˆë©´ 800 600 
        return image.resize((800, 600))
    

def image_rotate(image): # 180 ë„ íšŒì „ ì‹œì¼œì„œ ë°˜í™˜    
    return image.rotate(180)

def image_change_bw(image):
    return image.convert('L')
```
