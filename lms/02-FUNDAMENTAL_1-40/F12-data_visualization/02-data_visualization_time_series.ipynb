```python
# libary import
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import pandas as pd 
import seaborn as sns

import os

# font set
font_path = '/home/aiffel0042/anaconda3/envs/aiffel/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=18)
```

```python
# data import 

csv_path = "./flights.csv"
data = pd.read_csv(csv_path)  # 판다스 csv 읽기 메서드로 읽어오기
flighs = pd.DataFrame(data) # 판다스 데이터 프레임 객체로 ...
flighs
```

## graph 그리기

```python
sns.barplot(data=flighs, x="year", y="passengers")
```

```python
sns.pointplot(data=flighs, x="year", y="passengers")
```

```python
sns.lineplot(data=flighs, x="year", y="passengers")
```

```python
##hue 인자 
sns.lineplot(data=flighs, x="year", y="passengers", hue="month", palette='ch:.50')
plt.legend(bbox_to_anchor=(1.03,1), loc=2) # legend 그래프 밖에 추가하기? 범례
```

```python
# 히스토그램
sns.distplot(flighs["passengers"])
```
