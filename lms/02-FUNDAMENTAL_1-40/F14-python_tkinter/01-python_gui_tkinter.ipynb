```python
from tkinter import *
import tkinter.ttk as ttk

print('ì½”ë“œ ì‹¤í–‰ ì¤‘..ğŸ›¸')
window = Tk()   
window.title("ex3_Button")   
window.geometry("1024x512")     # ìœˆë„ìš° í¬ê¸°ë¥¼ ì¢€ë” í‚¤ì›ì‹œë‹¤.

label1 = ttk.Label(window, text="Label")
label1.grid(column=0, row=0)

combo = ttk.Combobox(window, width=10)
combo['values']= ("A", "B", "C")
combo.current(0) #set the selected item
combo.grid(row=1, column=1)

# í…ìŠ¤íŠ¸ ë°•ìŠ¤ 2ê°œ ìƒì„±
text1 = Text(window, width=60, height=40)
text2 = Text(window, width=60, height=40)
text1.grid(row=2, column=0, padx=20, pady=10)   
text2.grid(row=2, column=1, padx=20, pady=10)

# ë²„íŠ¼í´ë¦­ ì´ë²¤íŠ¸ í•¸ë“¤ëŸ¬ ì •ì˜
def click():
    label1.configure(text='ë²„íŠ¼ì´ ëˆŒë ¸ìŠµë‹ˆë‹¤.')   
    text1_string = text1.get("1.0","end")   # text1ì˜ í…ìŠ¤íŠ¸ê°’ì„ ê°€ì ¸ì™€ì„œ
    # 1.0 ì‹œì‘ë¬¸ì, "end " 
    text2_string = "{} and {}".format(text1_string, combo.get()) # text2ì— ë“¤ì–´ê°ˆ í…ìŠ¤íŠ¸ê°’ì„ ìƒì„±í•œ í›„
    text2.delete(1.0, "end")     # text2ì˜ í…ìŠ¤íŠ¸ê°’ì„ ì´ˆê¸°í™”í•˜ê³  ë‚˜ì„œ 
    text2.insert("end", text2_string)   # text2ì— ìƒˆë¡œ ìƒì„±í•œ í…ìŠ¤íŠ¸ë¥¼ ë„£ì–´ì¤ë‹ˆë‹¤.  

action = ttk.Button(window, text="ë²„íŠ¼ì„ ëˆŒëŸ¬ì£¼ì„¸ìš”", command=click)   
action.grid(row=0, column=1)

window.mainloop()
print('ì™„ë£Œ!')
```

## êµ¬ê¸€ë²ˆì—­ê¸° ë¼ì´ë¸ŒëŸ¬ë¦¬ í™œìš©

```python
from tkinter import *
import tkinter.ttk as ttk
from googletrans import Translator

print('ì½”ë“œ ì‹¤í–‰ ì¤‘..ğŸ›¸')
window = Tk()
window.title("ë‚˜ì˜ êµ¬ê¸€ ë²ˆì—­ê¸°")
window.geometry("1024x512")

# ë¼ë²¨ ìƒì„±
L1 = Label(window, text="ë²ˆì—­ ì „")
L2 = Label(window, text="ë²ˆì—­ í›„")
L1.grid(row=0, column=0)
L2.grid(row=0, column=2)


# ì½¤ë³´ë°•ìŠ¤ ìƒì„±
CB = ttk.Combobox(window)
CB['values']=("en","ko","ja")
CB.current(0)
CB.grid(row=0, column=1)

# í…ìŠ¤íŠ¸ ë°•ìŠ¤ ìƒì„±
T1 = Text(window)
T2 = Text(window)

T1.grid(row =1, column =0, padx=20, pady=10)
T2.grid(row =1, column =2, padx=20, pady=10)


# ë²ˆì—­ê¸° í•¨ìˆ˜
def translate(input_str, lang='en'):
    translator = Translator()
    result = translator.translate(input_str, dest=lang)
    return result.text

# ë²„íŠ¼í´ë¦­ ì½œë°± í•¨ìˆ˜    
def btnClick():
  
  b_str = T1.get(1.0, "end")# t1 í…ìŠ¤íŠ¸ ì¡°íšŒ
  cb_value = CB.get()
  
  a_str = translate(b_str, lang= cb_value) # ë²ˆì—­ê¸° ëŒë¦¬ê³  ê°’ ë°›ê¸°
  
  T2.delete(1.0, "end")# t2 ë¹„ìš°ê¸°
  T2.insert("end", a_str) # t2 ë²ˆì—­í•œ í…ìŠ¤íŠ¸ ì£¼ì…
  

# ë²„íŠ¼ ìƒì„±    
tBtn = ttk.Button(window, text="ë²ˆì—­", command = btnClick)
tBtn.grid(row=1, column=1)

window.mainloop()
print('ì™„ë£Œ!')
```
